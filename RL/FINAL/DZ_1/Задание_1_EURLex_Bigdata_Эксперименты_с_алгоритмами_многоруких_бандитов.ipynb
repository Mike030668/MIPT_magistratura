## Многорукие бандиты


Названия «однорукие бандиты» пошло из казино — так называются игровые автоматы с рычагом, за который можно потянуть и получить выигрыш.

Представьте, что вы находитесь в зале с десятком таких автоматов, и у вас есть N бесплатных попыток для игры на любом из них. Конечно, вы хотите выиграть побольше денег, но заранее не знаете, какой автомат дает самый большой выигрыш. Как раз цель многоруких бандитов — найти самый выгодный автомат с минимальными потерями.

```
import gym
import numpy as np
from gym.utils import seeding
from gym.vector.utils import spaces


class BanditEnv(gym.Env):
    """
    Bandit environment base to allow agents to interact with the class n-armed bandit
    in different variations
    p_dist:
        A list of probabilities of the likelihood that a particular bandit will pay out
    r_dist:
        A list of either rewards (if number) or means and standard deviations (if list)
        of the payout that bandit has
    info:
        Info about the environment that the agents is not supposed to know. For instance,
        info can releal the index of the optimal arm, or the value of prior parameter.
        Can be useful to evaluate the agent's perfomance
    """

    def __init__(self, p_dist, r_dist, info={}):
        if len(p_dist) != len(r_dist):
            raise ValueError("Probability and Reward distribution must be the same length")

        if min(p_dist) < 0 or max(p_dist) > 1:
            raise ValueError("All probabilities must be between 0 and 1")

        for reward in r_dist:
            if isinstance(reward, list) and reward[1] <= 0:
                raise ValueError("Standard deviation in rewards must all be greater than 0")

        self.p_dist = p_dist
        self.r_dist = r_dist

        self.info = info

        self.n_bandits = len(p_dist)
        self.action_space = spaces.Discrete(self.n_bandits)
        self.observation_space = spaces.Box(-1.0, 1.0, (1,))  #
        self.np_random = None
        self.seed()
        self.q_star = max([x[0] for x in self.r_dist])

    def seed(self, seed=None):
        self.np_random, seed = seeding.np_random(seed)
        return [seed]

    def step(self, action):
        assert self.action_space.contains(action)

        reward = 0
        done = True

        if np.random.uniform() < self.p_dist[action]:
            if not isinstance(self.r_dist[action], list):
                reward = self.r_dist[action]
            else:
                reward = np.random.normal(self.r_dist[action][0], self.r_dist[action][1])

        info = {"regret": self.q_star - self.r_dist[action][0]}
        return [0], reward, done, info  #

    def reset(self):
        return [0]  #

    def render(self, mode='human', close=False):
        pass


class BanditTenArmedGaussian(BanditEnv):
    """
    10 armed bandit mentioned on page 30 of Sutton and Barto's
    [Reinforcement Learning: An Introduction](https://www.dropbox.com/s/b3psxv2r0ccmf80/book2015oct.pdf?dl=0)
    Actions always pay out
    Mean of payout is pulled from a normal distribution (0, 1) (called q*(a))
    Actual reward is drawn from a normal distribution (q*(a), 1)
    """

    def __init__(self, bandits=10):
        p_dist = np.full(bandits, 1)
        r_dist = []

        for i in range(bandits):
            r_dist.append([np.random.normal(0, 1), 1])

        BanditEnv.__init__(self, p_dist=p_dist, r_dist=r_dist)
```

```
env = BanditTenArmedGaussian()

num_steps = 10000
q, r = 0, 0
for _ in range(num_steps):
    _, reward, _, info = env.step(env.action_space.sample())
    q += info['regret']
    r += reward

print(q / num_steps, r / num_steps)

```

## Краткое напоминание теории

Оптимальная полезность $V^*$: $$V^* = Q(a^*) = \max_{a \in A} Q(a)$$.

Regret (потери) — упущенное на каждом шаге вознаграждение: $$l_t=\mathbb{E}(V^*-Q(a_t)$$

Обновление полезностей:
$$Q_t(a) = Q_{t-1} + \frac{1}{N_t(a_t)}(r_t - Q_{t-1})$$

Алгоритм Upper Confidence Bound (UCB):
$$a_t=\textit{arg}\max_{a \in A}(Q(a) + \sqrt{\frac{c \log{t}}{N_t(a)}}$$

#### $\epsilon-greedy$ - жадной стратегии

Выбирает случайный автомат с вероятностью $\epsilon$ (изучение).

```
import numpy as np
import gym
from gym.vector.utils import spaces

class EpsilonGreedyAgent:
    def __init__(self, n_bandits, epsilon=0.1):
        self.n_bandits = n_bandits
        self.epsilon = epsilon
        self.q_values = np.zeros(n_bandits)
        self.action_counts = np.zeros(n_bandits)

    def select_action(self):
        if np.random.rand() < self.epsilon:
            return np.random.randint(self.n_bandits)
        else:
            return np.argmax(self.q_values)

    def update(self, action, reward):
        self.action_counts[action] += 1
        self.q_values[action] += (reward - self.q_values[action]) / self.action_counts[action]
```

```
env = BanditTenArmedGaussian()
agent = EpsilonGreedyAgent(n_bandits=10, epsilon=0.1)

num_steps = 10000
total_reward = 0
for _ in range(num_steps):
    action = agent.select_action()
    _, reward, _, _ = env.step(action)
    agent.update(action, reward)
    total_reward += reward

print(f"Среднее вознаграждение: {total_reward / num_steps}")
```

#### $\epsilon-greedy$ - жадной стратегии с затуханием

Выбирает лучший автомат по текущим оценкам с вероятностью 1-ε

```
class EpsilonGreedyDecayAgent:
    def __init__(self, n_bandits, initial_epsilon=1.0, decay_rate=0.999):
        self.n_bandits = n_bandits
        self.epsilon = initial_epsilon
        self.decay_rate = decay_rate
        self.q_values = np.zeros(n_bandits)
        self.action_counts = np.zeros(n_bandits)

    def select_action(self):
        if np.random.rand() < self.epsilon:
            return np.random.randint(self.n_bandits)
        else:
            return np.argmax(self.q_values)

    def update(self, action, reward):
        self.epsilon *= self.decay_rate
        self.action_counts[action] += 1
        self.q_values[action] += (reward - self.q_values[action]) / self.action_counts[action]
```

```
agent = EpsilonGreedyDecayAgent(n_bandits=10, initial_epsilon=1.0, decay_rate=0.999)

total_reward = 0
for _ in range(num_steps):
    action = agent.select_action()
    _, reward, _, _ = env.step(action)
    agent.update(action, reward)
    total_reward += reward

print(f"Среднее вознаграждение: {total_reward / num_steps}")
```

#### UCB (Upper Confidence Bound)

Выбирает автомат, который имеет наивысшую верхнюю границу доверительного интервала для среднего выигрыша

```
class UCBAgent:
    def __init__(self, n_bandits, c=2):
        self.n_bandits = n_bandits
        self.c = c
        self.q_values = np.zeros(n_bandits)
        self.action_counts = np.zeros(n_bandits)
        self.total_count = 0

    def select_action(self):
        if 0 in self.action_counts:
            return np.argmin(self.action_counts)
        ucb_values = self.q_values + self.c * np.sqrt(np.log(self.total_count) / self.action_counts)
        return np.argmax(ucb_values)

    def update(self, action, reward):
        self.total_count += 1
        self.action_counts[action] += 1
        self.q_values[action] += (reward - self.q_values[action]) / self.action_counts[action]
```

```
agent = UCBAgent(n_bandits=10, c=2)

total_reward = 0
for _ in range(num_steps):
    action = agent.select_action()
    _, reward, _, _ = env.step(action)
    agent.update(action, reward)
    total_reward += reward

print(f"Среднее вознаграждение: {total_reward / num_steps}")
```

#### function of experiments

Эта функция выполняет эксперименты, выбирает действия агентов, обновляет их состояния и логирует средние значения вознаграждений и потерь с использованием wandb.

```
def make_experiment(env, agent, num_steps=10000, agent_name="agent"):
    total_reward = 0
    total_regret = 0
    for step in range(num_steps):
        action = agent.select_action()
        _, reward, _, info = env.step(action)
        agent.update(action, reward)
        total_reward += reward
        total_regret += info['regret']
        # wandb.log
        wandb.log({f"{agent_name}_reward": total_reward / (step + 1),
                   f"{agent_name}_regret": total_regret / (step + 1)})
    return total_reward / num_steps, total_regret / num_steps
```

```
# Сброс окружения
env.reset()

# Тестирование агента UCB
ucb_agent = UCBAgent(n_bandits=10, c=2)
reward, regret = make_experiment(env, ucb_agent, agent_name="ucb")
print("Среднее вознаграждение (UCB):", reward)
print("Средняя потеря (UCB):", regret)
```

## Контекстные бандиты (Online Contextual Bandits)

### 1. Описание задачи

Контекстные бандиты служат для решения проблемы, аналогичной многоруким бандитам, с той лишь разницей, что на каждой итерации выбора действия есть дополнительная информация.

Задача может быть описана итеративным процессом получения данных:

* На каждой итерации среда генерирует наблюдение, которое состоит из вектора признаков фиксированного размера, и значение вознаграждения, который является стохастичным, но зависит от первого вектора.

* Агент должен выбрать одно действие из множества действий («рук») $m$.

* Среда сообщает вознаграждение для руки, которую выбрал агент.

Цель — построить стратегию, которая будет максимизировать вознаграждение, получаемое агентом.

Примерами сценариев использования могут быть задачи онлайн-рекомендаций, в которых указана лишь информация о том, нажал ли пользователь на рекламную ссылку или нет. Информации про другие ссылки, на которые он нажал, — нет. Или задачи клинических испытаний, в которых мы знаем, как человек отреагировал на одно лекарство, но не знаем, как отреагировал бы на другое.

### 2. Алгоритмы

В этом пункте мы частично рассмотрим библиотеку [Contextual Bandits](https://github.com/david-cortes/contextualbandits). Она предлагает реализации большого списка алгоритмов:

* Upper-Confidence Bounds: `BootstrappedUCB`, `LogisticUCB`, и `LinUCB`.

* Thompson Sampling: `BootstrappedTS`, `LogisticTS`, `LinTS`, and `SoftmaxExplorer`.

* Greedy Exploration: `EpsilonGreedy`.

* Adaptive Exploration: `AdaptiveGreedy`.

Более полную информацию вы можете найти [в документации](https://contextual-bandits.readthedocs.io/en/latest/).
Также обратите внимание [на примеры](https://github.com/david-cortes/contextualbandits/tree/master/example).

  - Цель: Построить и протестировать алгоритмы контекстных бандитов, чтобы определить, какой алгоритм лучше всего подходит для максимизации вознаграждения в данной задаче.

### Считывание данных

Данные представляют собой текстовый файл, используемый в LIBSVM и других системах, но имеющий первую строку. Она указывает на количество строк, столбцов и классов:

```
import numpy as np
import re
from scipy.sparse import lil_matrix
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.datasets import load_svmlight_file
from sklearn.preprocessing import StandardScaler

# Загрузка всех меток из полного набора данных
def load_all_labels(filename):
    labels = []
    with open(filename, "r") as f:
        f.readline()  # Пропустить первую строку с информацией
        for line in f:
            if not line.strip():
                continue
            label, _ = line.split(' ', 1)
            label = list(map(int, label.split(',')))
            labels.append(label)
    return labels


filename = "/content/EURLex-4.3K/train.txt"
# Загрузка всех меток
all_labels = load_all_labels(filename)

# Инициализация MultiLabelBinarizer со всеми возможными классами
mlb = MultiLabelBinarizer(classes=np.unique(np.hstack(all_labels)))
mlb.fit(all_labels)

# Загрузка и парсинг данных
def parse_data(filename, n_lines=None, mlb=None):
    with open(filename, "r") as f:
        infoline = f.readline().strip()
        n_features = int(re.sub(r"^\d+\s(\d+)\s\d+.*$", r"\1", infoline))

        if n_lines:
            data = []
            labels = []
            for i in range(n_lines):
                line = f.readline().strip()
                if not line:
                    break
                label, features = line.split(' ', 1)
                label = list(map(int, label.split(',')))
                features = features.split(' ')
                features = {int(feat.split(':')[0]): float(feat.split(':')[1]) for feat in features}
                data.append(features)
                labels.append(label)

            # Создание разреженной матрицы признаков
            X = lil_matrix((len(data), n_features))
            for i, feats in enumerate(data):
                for j, value in feats.items():
                    X[i, j] = value
            X = X.tocsr()

            # Преобразование разреженной матрицы в плотную
            X = X.toarray()

            # Преобразование меток с использованием всех возможных меток
            y = mlb.transform(labels)
        else:
            features, labels = load_svmlight_file(f, n_features=n_features, multilabel=True)
            labels = mlb.transform(labels)
            features = np.array(features.todense())
            features = np.ascontiguousarray(features)
            return features, labels

    return X, y

# Пример использования функции для частичного набора данных
X_, y_ = parse_data(filename, n_lines=8010, mlb=mlb)  # Загружаем первые 100 строк

# Масштабирование данных
scaler = StandardScaler()
X_ = scaler.fit_transform(X_)

sel_x, sel_y = 2000, 160
X, y = X_[:,:sel_x], y_[:,:sel_y]
print(X.shape)
print(y.shape)
```

## Пакетные модели (описание)

Логистическая регрессия и контекстные бандиты:

- `Bootstrapped Upper Confidence Bound` (C.I.=80%) - Использует метод бутстрапа для оценки доверительных интервалов и выбора действий.
- `Bootstrapped Thompson Sampling` - Бутстрапирование для метода Thompson Sampling.
- `Separate Classifiers + Beta Prior` - Отдельные классификаторы для каждого действия с априорным распределением Бета.
- `Epsilon-Greedy` (p0=20%, decay=0.9999) - Жадный алгоритм с вероятностью случайного выбора, которая постепенно уменьшается.
- `Logistic Upper Confidence Bound` (C.I.=70%) - Логистическая регрессия с оценкой доверительных интервалов.
- `Adaptive Greedy (decaying threshold)` - Адаптивная жадная стратегия с уменьшающимся порогом.
- `Adaptive Greedy (p0=30%, decaying percentile)`- Адаптивная жадная стратегия с уменьшающимся перцентилем.
- `Explore First (n=1,500)` - Сначала исследует действия, затем использует полученные знания.
- `Active Explorer` - Активное обучение для выбора действий.
- `Softmax Explorer` - Использует softmax-функцию для выбора действий.
- `Adaptive Active Greedy` - Адаптивная активная жадная стратегия.
- `Overall Best Arm (no context)` - Лучшее действие без учета контекста (контрольная линия).

```
from sklearn.linear_model import LogisticRegression

from contextualbandits.online import BootstrappedUCB, BootstrappedTS, LogisticUCB, \
            SeparateClassifiers, EpsilonGreedy, AdaptiveGreedy, ExploreFirst, \
            ActiveExplorer, SoftmaxExplorer
from copy import deepcopy

# Инициализация моделей
nchoices = y.shape[1]

# base_algorithm - играет ключевую роль в тестах бандитов, так как он определяет, как контекстные данные будут использоваться для оценки вероятностей успеха различных действий.
#base_algorithm = LogisticRegression(solver='lbfgs', warm_start=True)
base_algorithm = LogisticRegression(solver='lbfgs', max_iter=1000, warm_start=True)

# априорные распределения
beta_prior = ((3./nchoices, 4), 2) # базовое
beta_prior_ucb = ((5./nchoices, 4), 2) # Используется для UCB, и задает более оптимистичное априорное распределение для ускорения начального обучения
beta_prior_ts = ((2./np.log2(nchoices), 4), 2) # Используется для Thompson Sampling и задает консервативное распределение для учета большого количества действий.


## base algorithm в различных многоруких_бандитах

# BootstrappedUCB и BootstrappedTS базовые алгоритмы используются для обучения нескольких "bootstrap" моделей, каждая из которых обучается на случайной подвыборке данных.
bootstrapped_ucb = BootstrappedUCB(deepcopy(base_algorithm), nchoices = nchoices,
                                   beta_prior = beta_prior_ucb, percentile = 80,
                                   random_state = 1111)

bootstrapped_ts = BootstrappedTS(deepcopy(base_algorithm), nchoices = nchoices,
                                 beta_prior = beta_prior_ts, random_state = 2222)

# SeparateClassifiers алгоритм использует отдельный классификатор для каждого действия. Это полезно, когда каждое действие может иметь свои уникальные признаки, и мы хотим тренировать независимые модели для каждого действия.
one_vs_rest = SeparateClassifiers(deepcopy(base_algorithm), nchoices = nchoices,
                                  beta_prior = beta_prior, random_state = 3333)

# Для EpsilonGreedy базовый алгоритм используется для обучения модели, которая оценивает вероятности успеха действий. На основе этих оценок выбираются действия с определенной вероятностью случайного выбора ϵ
epsilon_greedy = EpsilonGreedy(deepcopy(base_algorithm), nchoices = nchoices,
                               beta_prior = beta_prior, random_state = 4444)

# LogisticUCB использует логистическую регрессию для оценки вероятностей успеха и выбора действий на основе этих оценок с применением UCB.
logistic_ucb = LogisticUCB(nchoices = nchoices, percentile = 70,
                           beta_prior = beta_prior_ts, random_state = 5555)

# В AdaptiveGreedy базовый алгоритм также используется для обучения модели и выбора действий с изменяющейся вероятностью случайного выбора 𝜖, которая адаптируется по мере обучения.
adaptive_greedy_thr = AdaptiveGreedy(deepcopy(base_algorithm), nchoices=nchoices,
                                     decay_type='threshold',
                                     beta_prior = beta_prior, random_state = 6666)

adaptive_greedy_perc = AdaptiveGreedy(deepcopy(base_algorithm), nchoices = nchoices,
                                      decay_type='percentile', decay=0.9997,
                                       beta_prior=beta_prior, random_state = 7777)

adaptive_active_greedy = AdaptiveGreedy(deepcopy(base_algorithm), nchoices = nchoices,
                                        active_choice='weighted', decay_type='percentile', decay=0.9997,
                                        beta_prior=beta_prior, random_state = 1234)

# ExploreFirst алгоритм сначала исследует действия на протяжении заданного числа итераций, а затем начинает использовать действия на основе своих оценок.
explore_first = ExploreFirst(deepcopy(base_algorithm), nchoices = nchoices,
                             explore_rounds=1500, beta_prior=None, random_state = 8888)

# ActiveExplorer алгоритм комбинирует активное обучение и алгоритмы контекстных бандитов, активно выбирая действия, которые максимизируют информацию.
active_explorer = ActiveExplorer(deepcopy(base_algorithm), nchoices = nchoices,
                                 beta_prior=beta_prior, random_state = 9999)

# SoftmaxExplorer алгоритм использует softmax-функцию для выбора действий. Вероятность выбора действия пропорциональна экспоненте его ожидаемого вознаграждения.
softmax_explorer = SoftmaxExplorer(deepcopy(base_algorithm), nchoices = nchoices,
                                   beta_prior=beta_prior, random_state = 5678)

# набор моделей
models = [bootstrapped_ucb,
          bootstrapped_ts, one_vs_rest, epsilon_greedy, logistic_ucb,
          adaptive_greedy_thr, adaptive_greedy_perc, adaptive_active_greedy,
          explore_first, active_explorer,  softmax_explorer
          ]
```

```
# Подготовка для сохранения результатов
rewards_ucb, rewards_ts, rewards_ovr, rewards_egr, rewards_lucb, rewards_agr, rewards_agr2, rewards_aag, rewards_efr, rewards_ae, rewards_sft = [list() for i in range(len(models))]

lst_rewards = [rewards_ucb, rewards_ts, rewards_ovr, rewards_egr, rewards_lucb, rewards_agr, rewards_agr2, rewards_aag, rewards_efr, rewards_ae, rewards_sft]

# batch size - алгоритмы будут обновлены после N раундов
batch_size = 50

# начальное начальное значение — все политики начинаются с одного и того же небольшого случайного выбора действий/наград.
first_batch = X[:batch_size, :]
np.random.seed(1)
action_chosen = np.random.randint(nchoices, size=batch_size)
rewards_received = y[np.arange(batch_size), action_chosen]

# fitting models for the first time
for model in models:
    model.fit(X=first_batch, a=action_chosen, r=rewards_received)

# these lists will keep track of which actions does each policy choose
lst_a_ucb, lst_a_ts, lst_a_ovr, lst_a_egr, lst_a_lucb, lst_a_agr, lst_a_agr2, lst_a_ag, lst_a_efr, lst_a_ae, lst_a_sft = [action_chosen.copy() for i in range(len(models))]

lst_actions = [lst_a_ucb, lst_a_ts, lst_a_ovr, lst_a_egr, lst_a_lucb, lst_a_agr, lst_a_agr2, lst_a_ag, lst_a_efr, lst_a_ae, lst_a_sft]

# раундов моделируются из полного набора данных
def simulate_rounds(model, rewards, actions_hist, X_global, y_global, batch_st, batch_end):
    np.random.seed(batch_st)

    # выбор действий для этого пакета
    actions_this_batch = model.predict(X_global[batch_st:batch_end, :]).astype('uint8')

    # отслеживание суммы полученных вознаграждений
    rewards.append(y_global[np.arange(batch_st, batch_end), actions_this_batch].sum())

    # добавляем этот пакет в историю выбранных действий
    new_actions_hist = np.append(actions_hist, actions_this_batch)

    # сейчас переоснащаю алгоритмы после наблюдения за этими новыми наградами
    np.random.seed(batch_st)
    model.fit(X_global[:batch_end, :], new_actions_hist, y_global[np.arange(batch_end), new_actions_hist])

    return new_actions_hist

# теперь запускаем всю симуляцию
for i in range(int(np.floor(X.shape[0] / batch_size))):
    batch_st = (i + 1) * batch_size
    batch_end = (i + 2) * batch_size
    batch_end = np.min([batch_end, X.shape[0]])

    for model in range(len(models)):
        lst_actions[model] = simulate_rounds(models[model],
                                             lst_rewards[model],
                                             lst_actions[model],
                                             X, y,
                                             batch_st, batch_end)
```

### Визуализация результатов

```
import matplotlib.pyplot as plt
from pylab import rcParams
%matplotlib inline

def get_mean_reward(reward_lst, batch_size=batch_size):
    mean_rew=list()
    for r in range(len(reward_lst)):
        mean_rew.append(sum(reward_lst[:r+1]) * 1.0 / ((r+1)*batch_size))
    return mean_rew

rcParams['figure.figsize'] = 25, 15
lwd = 5
cmap = plt.get_cmap('tab20')
colors=plt.cm.tab20(np.linspace(0, 1, 20))

ax = plt.subplot(111)
plt.plot(get_mean_reward(rewards_ucb), label="Bootstrapped Upper Confidence Bound (C.I.=80%)", linewidth=lwd,color=colors[0])
plt.plot(get_mean_reward(rewards_ts), label="Bootstrapped Thompson Sampling",linewidth=lwd,color=colors[2])
plt.plot(get_mean_reward(rewards_ovr), label="Separate Classifiers + Beta Prior",linewidth=lwd,color=colors[4])
plt.plot(get_mean_reward(rewards_egr), label="Epsilon-Greedy (p0=20%, decay=0.9999)",linewidth=lwd,color=colors[6])
plt.plot(get_mean_reward(rewards_lucb), label="Logistic Upper Confidence Bound (C.I.=70%)",linewidth=lwd,color=colors[8])
plt.plot(get_mean_reward(rewards_agr), label="Adaptive Greedy (decaying threshold)",linewidth=lwd,color=colors[10])
plt.plot(get_mean_reward(rewards_agr2), label="Adaptive Greedy (p0=30%, decaying percentile)",linewidth=lwd,color=colors[12])
plt.plot(get_mean_reward(rewards_aag), label="Adaptive Active Greedy", linewidth=lwd,color=colors[16])
plt.plot(get_mean_reward(rewards_efr), label="Explore First (n=1,500)",linewidth=lwd,color=colors[14])
plt.plot(get_mean_reward(rewards_ae), label="Active Explorer", linewidth=lwd,color=colors[15])
plt.plot(get_mean_reward(rewards_sft), label="Softmax Explorer", linewidth=lwd,color=colors[17])
plt.plot(np.repeat(y.mean(axis=0).max(), len(rewards_ucb)), label="Overall Best Arm (no context)", linewidth=lwd,color=colors[1],ls='dashed')


# import warnings
box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1,
                 box.width, box.height * 1.25])
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, ncol=3, prop={'size':20})


plt.tick_params(axis='both', which='major', labelsize=25)
plt.xticks([i*20 for i in range(9)], [i*1000 for i in range(9)])


plt.xlabel('Rounds (models were updated every 50 rounds)', size=30)
plt.ylabel('Cumulative Mean Reward', size=30)
plt.title('Comparison of Online Contextual Bandit Policies\n(Base Algorithm is Logistic Regression)\n\nEURLex_Big Dataset\n(160 categories, 2000 attributes)',size=30)
plt.grid()
plt.show()
```
